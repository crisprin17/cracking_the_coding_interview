```python
# define a Tree Class

class binary_tree(object):
 
    def __init__(self, valore):
        # inizializza il nodo
        self.nodo = valore
        #inizializza figlio desto e sinistro a None
        self.figlioSinistro = None
        self.figlioDestro = None

        
    def insert_left(self, valore):
        self.figlioSinistro = binary_tree(valore)
        return self.figlioSinistro
    
    
    def insert_right(self, valore):
        self.figlioDestro = binary_tree(valore)
        return self.figlioDestro
    
```

```python
def check_node_binary_search_tree(nodo, valore):
    print(nodo.nodo)
    if not nodo.nodo:  
        return None
  
    if (nodo.nodo == valore):  
        print('value found')
        return nodo.nodo
    
    
    if nodo.figlioSinistro:
        recursive_left = check_node_binary_search_tree(nodo.figlioSinistro, valore)  
        if recursive_left:
            return True # node found, no need to look further 
        else:
            print('value not found left')
    
    if nodo.figlioDestro:
        recursive_right = check_node_binary_search_tree(nodo.figlioDestro, valore)  
        if recursive_right:
            return True
        else:
            print('value not found right')
```

```python
def minValueNode(nodo): 
    current = nodo
    while(current.figlioSinistro is not None): 
        current = current.figlioSinistro
    print ('return current', current.nodo)
    return current
  
def deleteNode(nodo, valore): 
    
    if nodo is None: 
        return nodo  
    
    if valore < nodo.nodo:
        nodo.figlioSinistro = deleteNode(nodo.figlioSinistro, valore) 
  

    if valore > nodo.nodo: 
        nodo.figlioDestro = deleteNode(nodo.figlioDestro, valore) 
  
    else: 
        if nodo.figlioSinistro is None : 
            temp = nodo.figlioDestro  
            root = None 
            return temp  
              
        elif nodo.figlioDestro is None : 
            temp = nodo.figlioSinistro  
            root = None
            return temp 
  
        # Node with two children: Get the inorder successor 
        # (smallest in the right subtree) 
        temp = minValueNode(nodo.figlioSinistro) 
  
        # Copy the inorder successor's content to this node 
        nodo.nodo = temp.nodo
  
        # Delete the inorder successor 
        nodo.figlioSinistro = deleteNode(nodo.figlioSinistro, temp.nodo) 
  

    return nodo
```

```python
def insert_binary_node(nodo, valore): 
    if nodo is None:
        return nodo
    
    if (nodo.figlioDestro is None) and (nodo.figlioSinistro is None): 
        if valore > nodo.nodo:
            nodo.figlioDestro = binary_tree.insert_right(nodo, valore)
        elif valore < nodo.nodo:
            nodo.figlioSinistro = binary_tree.insert_left(nodo, valore)
        
    
    elif valore < nodo.nodo:
        nodo.figlioSinistro = insert_binary_node(nodo.figlioSinistro, valore) 
  
    elif valore > nodo.nodo: 
        nodo.figlioDestro = insert_binary_node(nodo.figlioDestro, valore) 
    return nodo  
```

```python
def print_inorder_tree(nodo):
    if (nodo):
        print(nodo.nodo)
        print_inorder_tree(nodo.figlioSinistro)
        print_inorder_tree(nodo.figlioDestro)
```
